## Problem Statement: In the agricultural sector, accurately forecasting crop yields is crucial for ensuring food security and optimizing resource allocation. This project addresses the challenge of predicting crop yields for Kharif season crops in regions with similar characteristics. Using a dataset containing attributes such as climatic conditions, soil characteristics, air quality, and agricultural inputs (e.g., fertilizers and pesticides), exploratory data analysis (EDA) was conducted to uncover key patterns. A time-series machine learning or deep learning model was then developed to predict crop yields, enabling better planning and decision-making for sustainable agricultural practices.

## ARIMA Forecasting for next 3 periods: (High Yield)

```python
from pmdarima import auto_arima
```

```python
arima= auto_arima(df_kharif_High['Yield'],trace=True,seasonal=True)
```

```python
arima
```

```python
arima.summary()
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
## Train-test split:
```

```python
df_kharif_High_i= df_kharif_High.reset_index(inplace=True,drop=True)
df_kharif_Low_i= df_kharif_Low.reset_index(inplace=True,drop=True)
df_kharif_Mid_i= df_kharif_Mid.reset_index(inplace=True,drop=True)
```

```python
df_kharif_High
```

```python
df_kharif_Mid
```

```python
df_kharif_Low
```

```python
train_high= df_kharif_High['Yield'].iloc[0:20]
```

```python
train_high
```

```python
test_high=df_kharif_High['Yield'].iloc[20:]
```

```python
test_high
```

```python
arima_model= ARIMA(train_high,order=(1,1,0))
```

```python
arima_model
```

```python
arima_model_fit= arima_model.fit()
```

```python
arima_model_fit # Model is fit and ready to forecast
```

```python
# prediction :
prediction=arima_model_fit.predict(start= len(train_high),                    
                                    end=len(train_high)+len(test_high)+ 2, 
                                    dynamic=True)
```

```python
prediction
```

```python
test_high
```

```python
plt.figure(figsize=(20,10))
df_kharif_High['Yield'].plot()
prediction.plot()
plt.grid()
```
